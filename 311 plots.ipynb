```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.interpolate import griddata
```

```python
training_data = pandas.read_csv('/home/mike/311/data/train.csv')
```

```python
city_boundaries = [(37.4,37.7),(37.7,37.9),(41.24,41.3),(41.3,41.4),(41.6,42.1)]

def select(data,city):
    """
    Return a data frame for the city within the correct latitudes
    """
    latitude_min = city[0]
    latitude_max = city[1]
    selected = data[(data.latitude >= latitude_min) & (data.latitude < latitude_max)]
    return selected

cities = [select(training_data,boundary) for boundary in city_boundaries]
```

```python
def plot_city(city_num):
    x,y = cities[city_num].latitude.values,cities[city_num].longitude.values
    
    heatmap, xedges, yedges = np.histogram2d(x, y, bins=30)
    extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

    plt.clf()
    plt.imshow(heatmap, extent=extent)
    plt.colorbar()
    plt.rcParams['figure.figsize'] = (10,10)  # 10x6 inches 
    plt.show()
        
    z = cities[city_num].num_views.values
    
    xi=np.linspace(x.min(),x.max(),20)
    yi=np.linspace(y.min(),y.max(),20)

    X,Y= np.meshgrid(xi,yi)
    Z = griddata((x, y), z, (X, Y))
    plt.rcParams['figure.figsize'] = (10,10)  # 10x6 inches 
    plt.contourf(X,Y,Z,method='cubic')
    plt.colorbar()

    plt.show()
    
    plt.plot(x,z,'x')
    plt.show()
    plt.plot(y,z,'x')
```

```python
plot_city(3)
```

```python
def in_box(box,point):
    lat_extent = box[0]
    lon_extent = box[1]
    
    in_extent = lambda g,extent : extent[0] < g < extent[1]
    inside_box = lambda box,point : in_extent(point[0],box[0]) and in_extent(point[1],box[1])
        
    return inside_box(box,point)
```

```python
box = ((0,1),(0,2))
point = (2.5,2.5)
in_box(box,point)
```

```python
def which_box(boxes,point):
    i = -1
    for box_num,box in enumerate(boxes):
        if in_box(box,point):
            i = box_num
    
    return i
```

```python
box_0 = ((0,1),(0,1))
box_1 = ((0,2),(0,2))
box_2 = ((2,3),(2,3))

boxes = [box_1,box_0,box_2] #arranged hierarchically

point = (2.5,2.5)

which_box(boxes,point)
```
